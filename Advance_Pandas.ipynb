```
import pandas as pd
import numpy as np

invoices = pd.read_csv('https://raw.githubusercontent.com/payalkutana/python-pandas/main/invoices1.csv')
#print(invoices)
invoices.dtypes

#sample() used to generate a sample random from the function caller dataFrame
invoices.sample(5)  
```

# **Data Types and Conversion**

```
invoices.dtypes
```

```
#Conerting datatype using -> astype

invoices['Type of Meal']=invoices['Type of Meal'].astype('category')
invoices['Date'] = invoices['Date'].astype('datetime64')
invoices['Meal Price'] = invoices['Meal Price'].astype('int')

invoices.dtypes
```

```
#Converting datatype using conversion helper function -> pd.to_<some_type>

#pd.to_numeric()

invoices.loc[45,'Meal Price'] = 'I am causing trouble'
invoices.loc[35,'Meal Price'] = 'Me too'

invoices.dtypes
invoices['Meal Price'].apply(lambda x: type(x)).value_counts()
```

```
invoices['Meal Price'][invoices['Meal Price'].apply(
  lambda x: isinstance(x,str)
)]
```

```
invoices['Meal Price'].astype(int) #Because there is no obvious way to convert the string into an integer
```

```
# convert the offending values into np.nan
invoices['Meal Price'] = pd.to_numeric(invoices['Meal Price'],errors='coerce')
```

```
# fill np.nan with the median of the data
invoices['Meal Price'] = invoices['Meal Price'].fillna(invoices['Meal Price'].median())
```

```
# convert the column into integer
invoices['Meal Price'].astype(int)
```

```
# pd.to_datetime()    -> method converts a string into a datetime format

print(pd.to_datetime('2019-8-1'))
print(pd.to_datetime('2019/8/1'))
print(pd.to_datetime('8/1/2019'))
print(pd.to_datetime('Aug, 1 2019'))
print(pd.to_datetime('Aug - 1 2019'))
print(pd.to_datetime('August - 1 2019'))
print(pd.to_datetime('2019, August - 1'))
print(pd.to_datetime('20190108'))
```

```
#exact=False Pandas tries to match the pattern anywhere in the date string

print(pd.to_datetime('yolo 20190108', format='%Y%d%m', exact=False))  
```

# Concatenating 

**Concatenating - Vertical Stacking**

```
#use dt accessor to count all the years
invoices['Date of Meal'].dt.year.value_counts().sort_index()
```

```
#Split the data
# copy() to make sure that the resulting DataFrames are going to be a copy of the data and not just a reference to the original DataFrame.
y_2013 = invoices[invoices['Date of Meal'].dt.year == 2013].copy()
y_2014 = invoices[invoices['Date of Meal'].dt.year == 2014].copy()
y_2015 = invoices[invoices['Date of Meal'].dt.year == 2015].copy()
y_2016 = invoices[invoices['Date of Meal'].dt.year == 2016].copy()
y_2017 = invoices[invoices['Date of Meal'].dt.year == 2017].copy()
y_2018 = invoices[invoices['Date of Meal'].dt.year == 2018].copy()
y_2019 = invoices[invoices['Date of Meal'].dt.year == 2019].copy()
```

```
y_2014
```

```
# Concatenating data vertically
pd.concat([y_2013,y_2014,y_2015,y_2016,y_2017,y_2018,y_2019],axis=0,keys=['1','2','3','4','5','6','7'],names=['index','year_of_order'])
```

```
# Concatenating data Horizontally
import datetime

range_a = pd.date_range(
    datetime.datetime(2019,1,2),
    datetime.datetime(2019,1,8)
)
df_a = pd.DataFrame(
    index=range_a, 
    data=np.random.randint(2,10,size=len(range_a)),
    columns=['observations_A']
)
range_b = pd.date_range(
    datetime.datetime(2019,1,5),
    datetime.datetime(2019,1,12)
)
df_b = pd.DataFrame(
    index=range_b, 
    data=np.random.randint(2,10,size=len(range_b)),
    columns=['observations_B']
)
pd.concat([df_a,df_b],axis=1)
```

**Stack/Unstack**

```
temp=invoices.groupby(['Company Id','Type of Meal','Heroes Adjustment']).agg({'Meal Price':np.mean})
temp
```

```
temp.unstack(level='Type of Meal')
```

```
stack_test = invoices.groupby(['Company Id','Type of Meal']).agg({
    'Meal Price':[max,min,np.mean],
    'Date of Meal':[max,min],
})
stack_test
```

```
stack_test.stack()
```

**Melt**

```
melt_experiment = pd.merge(
    invoices,
    pd.get_dummies(invoices['Type of Meal']).mul(invoices['Meal Price'].values,axis=0),
    left_index=True,
    right_index=True
)
del melt_experiment['Type of Meal']
del melt_experiment['Meal Price']
melt_experiment
```

```
pd.melt(
    frame=melt_experiment,
    id_vars=['Order Id', 'Date', 'Meal Id', 'Company Id', 'Date of Meal','Participants', 'Heroes Adjustment'],
    value_vars=['Breakfast', 'Dinner', 'Lunch'],
    var_name='Type of Meal',
    value_name='Expenses'
)
```
